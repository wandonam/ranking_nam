```python
from datetime import datetime
from pathlib import Path
import pandas as pd
```

```python
# ================= NAVER =================
base_path = Path(r'..\00_Database\01_naver\01_healthyfood')

csv_files = sorted(base_path.glob('*.csv'), key=lambda x: x.stem, reverse=True)
latest_df = pd.read_csv(csv_files[0])
previous_df = pd.read_csv(csv_files[1])

prev = previous_df.set_index('code')

latest_df['prev_price'] = latest_df['code'].map(prev['price'])
latest_df['prev_review'] = latest_df['code'].map(prev['review'])
latest_df['prev_rank'] = latest_df['code'].map(prev['rank'])
latest_df['prev_star'] = latest_df['code'].map(prev['star'])

latest_df['price_diff'] = latest_df['price'] - latest_df['prev_price']
latest_df['review_diff'] = latest_df['review'] - latest_df['prev_review']
latest_df['rank_diff'] = latest_df['prev_rank'] - latest_df['rank']
latest_df['star_diff'] = latest_df['prev_star'] - latest_df['star']

latest_df.drop(columns=[col for col in latest_df.columns if col.startswith('prev_')], inplace=True)

naver_10 = latest_df[latest_df['rank'].between(1,10)]
naver_hot = latest_df[latest_df['rank_diff'] >= 10]
naver_down = latest_df[latest_df['rank_diff'] <= -10]
```
