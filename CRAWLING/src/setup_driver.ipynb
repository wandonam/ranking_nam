### <mark>**✅Setup_Driver**

##### **Chrome Driver**
* https://googlechromelabs.github.io/chrome-for-testing/#stable

```python
import platform
import sys, os, requests, zipfile
```

```python
os_name = platform.system().lower()
architecture = platform.machine()

if os_name == 'darwin':
    if architecture == 'arm64':
        print("운영체제: macOS, 아키텍처: ARM64")
    elif architecture == 'x86_64':
        print("운영체제: macOS, 아키텍처: x64")
elif os_name == 'windows':
    if sys.maxsize > 2**32:
        print("운영체제: Windows, 아키텍처: 64-bit")
    else:
        print("운영체제: Windows, 아키텍처: 32-bit")
else:
    print(f"운영체제: {os_name}, 아키텍처: {architecture}")
```

##### **Chrome version**

```python
chrome_driver_url = 'https://storage.googleapis.com/chrome-for-testing-public/146.0.7680.178/win64/chromedriver-win64.zip'
```

```python
# 다운로드
os.makedirs(rf'../02_Driver', exist_ok=True)
with requests.get(chrome_driver_url) as response:
    with open('../02_Driver/chromedriver.zip', 'wb') as file:
        file.write(response.content)


# 압축해제
with zipfile.ZipFile('../02_Driver/chromedriver.zip') as zip_ref:
    zip_ref.extractall('../02_Driver')
os.remove('../02_Driver/chromedriver.zip')
```
